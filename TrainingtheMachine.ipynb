```python
ndims = X_train.shape[1] * X_train.shape[2]
X_train = X_train.reshape(X_train.shape[0], ndims)
print("Training Shape:", X_train.shape)
```

```python
scaler = MinMaxScaler().fit(X_train)

X_train = scaler.transform(X_train)
```
